```python
print("People with and without glasses count : ", df1['glass'].nunique())
print("types: ", df1['glass'].unique())
pd.value_counts(df1['glass'])
```

```python
plt.figure(figsize = (8,8))
plt.pie(pd.value_counts(df1['glass']), labels = ['Normal','None','Dark'],autopct = '%.2f%%',textprops = {'size' : 'x-large','fontweight' : 'bold','color' : 'black'})
plt.legend()
plt.title('Types of specs worn in Percentage', fontsize = 18, fontweight = 'bold')
plt.show()
```
